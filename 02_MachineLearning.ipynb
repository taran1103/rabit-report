**②非線形回帰**

非線形回帰とは、複雑な非線形構造を内在する現象を非線形回帰モデリングを実施する。

基底関数を使用して、表現する。道パラメータには最小２乗法や最尤法で推定する。

学習データに対して、十分小さな誤差が得られないモデル（←未学習）や

小さな誤差は得られたがテスト集合誤差との差が大きいモデル（←過学習）があり、

モデルが煩雑のため、適切な表現にすることが難しい。



調整するために対策として以下３点があげられる。

①学習データ数の増加

②不要な関数（変数）の削除による表現力の抑止

③正則化法を利用し表現力の抑止

　　※正則化とは※

　　学習時に損失関数に正則化項を加えたものを目的関数とし、最適化実施。

　　これにより正則化項がパラメータの大きさに対するペナルティとして作用し、

　　抑制がかかることで過学習を防ぐ効果をもたらす。


他にも、ホールドアウト法やクロスバリデーション（交差検証）がある。

**■k-means**

教師なし学習でクラスタリング手法を用いて与えられたデータをk個のクラスタに分類

【算出方法】

①各クラスタ中心の初期値を設定

②各データ点に対して、各クラスタ中心との距離を計算し、最も距離が近いクラスタを割り当てる

③各クラスタの平均ベクトル（中心）を計算する

④収束するまで②③を繰り返す
